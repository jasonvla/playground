```python

import networkx as nx
import matplotlib.pyplot as plt
import requests
import os
base_dir = os.getcwd()
file_path = os.path.join(base_dir, "RadioheadX.graphml")
G = nx.read_graphml(file_path)
```

```python
# ==================================================
# CALCULATING THE AVG WEIGHTED DEGREE OF THE NETWORK
# ==================================================
all_nodes = G.nodes()
allweights = 0
for (node1,node2,data) in G.edges(data=True):
    edgeweights_list = [list(data.values())[0]]
    for weights in edgeweights_list:
        allweights += weights
avg_wDeg = allweights * 2 / len(all_nodes)
print(f"{avg_wDeg:.2f}")
```

```python
# =============================================================
# CALCULATING THE WEIGHTED DEGREE FOR ALL NODES IN THE NETWORK
# =============================================================

nodes_weights = G.degree(weight='weight')
nodes_weights = dict(nodes_weights)
for wDegs in nodes_weights:
    wDegs = nodes_weights.values()
weightedDegreesOfNodes = list(wDegs)
weightedDegreesOfNodes
```

```python
# ===================================================
# GETTING THE VIEWCOUNTS FOR ALL NODES IN THE NETWORK
# ===================================================

viewcountNodePairs = nx.get_node_attributes(G, "viewcount")
for views in viewcountNodePairs:
    views = viewcountNodePairs.values()
viewcounts = list(views)
viewcounts
```

```python
# ==================================================================
# CALCULATING THE CORRELATION COEFFICIENT BETWEEN VIEWCOUNT AND WDEG
# ==================================================================

from scipy.stats import pearsonr
corr = pearsonr(viewcounts, weightedDegreesOfNodes)
print(corr)
```
